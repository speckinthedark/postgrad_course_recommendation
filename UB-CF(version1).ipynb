```python
import pandas as pd
#load data in, this is the data from approach2
data = pd.read_csv('C:/Users/zentz/Desktop/project/postgrad_course_recommendation-master/student_db_app2_truncnorm.csv')
print(data.shape)
ug_courses = ["BOA","BOB","BOC","BOE","BSC"]
data = data.drop("Unnamed: 0",axis = 1)
data.head()
```

```python
#data preproces
#data["COURSE_CODE"] = data["COURSE_CODE"].fillna('not_taking')
data = data.dropna(subset=["COURSE_CODE"])
#we dont want the student who did not take PG course in the data
data = data[data.columns].fillna('0')
data = data.reset_index(drop = True)
print(data.shape)
data.head()
```

```python
#generate sublist
subject_course_list = []
for course in ug_courses:
    for i in range(0,16):
        subject_course_list += [course+'_subject_'+str(i+1)]
profile_title_list = data.columns.difference(subject_course_list)
print(profile_title_list)
print(subject_course_list)
```

```python
#data transformation
#in this step, apply feature selection 
y = pd.DataFrame(data,columns=["COURSE_CODE","CODE_DESCR_mast"])
X = data.drop(["COURSE_CODE","CODE_DESCR_mast"],axis=1)
print("Now the size of x is:",X.shape,"and the size of y is:",y.shape)
#feature seleciton
useless_profile_title_list = ['STU_ID_PROXY','UG_course','CODE_DESCR_bach','FACULTY_CODE_mast','LANGUAGE_TEST_SCORE','LANGUAGE_TEST_TYPE']+['FACULTY_CODE_mast']
# data cleaning, to sparse, or useless
useful_profile_title_list= list(set(profile_title_list)-set(useless_profile_title_list)-set(y.columns))
column_title_need_onehotEncode = ['COURSE_CREDITS_REQ_bach','FACULTY_CODE_bach','LOCATION_CODE_bach','top5_bach_bach',
                                  'AGE_BAND','STATE','HOME_LANGUAGE','ATAR_BAND',
                                  'COURSE_CREDITS_REQ_mast','LOCATION_CODE_mast','STUDY_YEAR_mast','N_SUBJECTS_mast']

X.isnull().sum().sum()
```

```python
#one hot encoder
from sklearn.preprocessing import LabelBinarizer
from sklearn.preprocessing import OneHotEncoder

####binary encoder
encoder = LabelBinarizer()
Y_values = encoder.fit_transform(y["COURSE_CODE"].values)

X_part1 = pd.DataFrame({})
for column in column_title_need_onehotEncode:
    column_values=X[column_title_need_onehotEncode][column].astype(str)
    binarised_x = encoder.fit_transform(column_values.values)
    X_part1 = pd.concat([X_part1,pd.DataFrame(binarised_x)],axis=1)

##### one hot encoder       
X_part2 = X[list(set(useful_profile_title_list)-set(column_title_need_onehotEncode))]
X_no_syn = pd.concat([X_part1,X_part2],axis = 1,ignore_index=True)
X_whole = pd.concat([X_no_syn,X[subject_course_list]],axis = 1)
X_values = X_whole.values
print("X_part1:",X_part1.shape)
print("X_part2:",X_part2.shape)
print("X_no_syn:",X_no_syn.shape)
print("X[useful_profile_title_list]:",X[useful_profile_title_list].shape)
print("X_whole:",X_whole.shape)
print("X_values:",X_values.shape)
print("Y_values:",Y_values.shape)
Y_label = y["COURSE_CODE"]
X_whole
```

```python
#start to build UB_CF system
#actually now the system is built in all data, can be used to check
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import timeit
start = timeit.default_timer()

b = cosine_similarity(X_whole)
#we want to find the biggest similarity, so the diagonal need to be set as 0
np.fill_diagonal(b, 0 )

similarity = pd.DataFrame(b)

stop = timeit.default_timer()
print('Time: ', stop - start)
similarity
```

```python
def find_n_neighbours(df,n):
    #order = np.argsort(df.values, axis=1)[:, :n]
    df = df.apply(lambda x: pd.Series(x.sort_values(ascending=False).iloc[:n].index, index=['top{}'.format(i) for i in range(1, n+1)]), axis=1)
    return df
```

```python
sim_student_30 = find_n_neighbours(similarity,30)
sim_student_30.head()
```

```python
#above system is built in train set, however we need to build it for test set
#split train test set
from sklearn.model_selection import train_test_split
import numpy as np
X_train, X_test, Y_train, Y_test = train_test_split(X_whole, Y_label, test_size=0.33, random_state=8)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
X_train = X_train.reset_index(drop = True)
X_test = X_test.reset_index(drop = True)
Y_train = Y_train.reset_index(drop = True)
Y_test = Y_test.reset_index(drop = True)
```

```python
#ignore clustering, use basis similarity
#we need to compare every user in test set to all user in train set
similarity_test_train = pd.DataFrame(cosine_similarity(X_test,X_train))
print(similarity_test_train.shape)
similarity_test_train.head()
```

```python
sim_student_30_test = find_n_neighbours(similarity_test_train,30)
print(sim_student_30_test.shape)
sim_student_30_test.head()
```

```python
#rough test
print(Y_test.iloc[1500])
print("--------------------------------")
for column in sim_student_30_test.iloc[1500]:
    print(Y_train.iloc[column])
```

```python
#check accuracy, so now we first need to generate no repeated top5 PG
top5_PG_course = []   #containing whole top 5 dataset
for student in range(sim_student_30_test.shape[0]):
    top5_list = []  #list for a student
    for top_n_similar_student in sim_student_30_test.iloc[student]:
        if Y_train[top_n_similar_student] not in top5_list and len(top5_list) < 5:
            top5_list.append(Y_train[top_n_similar_student])
        elif len(top5_list)>5: 
            break
    top5_PG_course.append(top5_list)
top5_PG_course = pd.DataFrame(top5_PG_course)
print(top5_PG_course.shape)
top5_PG_course.head()
```

```python
#generate a score
from sklearn import metrics
from sklearn.metrics import classification_report
Y_pre = []
right = 0
for i in range(Y_test.shape[0]):
    if Y_test.iloc[i] in list(top5_PG_course.iloc[i]):
        Y_pre.append(Y_test.iloc[i])
        right += 1
    else:
        Y_pre.append("NA")      
persicion_in_5 = right/Y_test.shape[0]
print("top5 weight f1 score:",metrics.f1_score(Y_test, Y_pre, average="weighted"))
print(classification_report(Y_test, Y_pre))
print(persicion_in_5)
```
